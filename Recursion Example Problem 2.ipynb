## Problem 2
Given an integer, create a function which returns the sum of all the individual digits in that integer. For example: if n = 4321, return (4+3+2+1) = 10.

Hints:

We'll need to think of this function recursively by knowing that: 4502 % 10 + sum_func(4502/10)

```python
def sum_func(n):
    
    if n == 0:
        return 0
    
    else:
        return (n % 10) + sum_func(n // 10)
    
    pass
```

```python
sum_func(4321)
```
