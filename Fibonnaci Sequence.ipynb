```python
def fib_iter(n):
    
    a, b = 0, 1
    
    for i in range(n):
        
        a, b = b, a + b
        
    return a
```

```python
fib_iter(10)
```

```python
def fib_rec(n):
    
    # Base Case
    
    if n == 0 or n == 1:
        return n
    
    # Recursive Case
    else:
        return fib_rec(n - 1) + fib_rec(n - 2)
```

```python
fib_rec(10)
```

```python
# Instantiate cache information
n = 30
cache = [None] * (n + 1)

def  fib_dyn(n):
    
    # Base Case
    if n == 0 or n == 1:
        return n
    
    # Check Cache
    if cache[n] != None:
        return cache[n]
    
    # Keep setting Cache
    cache[n] = fib_dyn(n - 1) + fib_dyn(n - 2)
    
    return cache[n]
```

```python
fib_dyn(10)
```
