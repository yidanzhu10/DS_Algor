```python
def permute(s):
    
    out = []
    
    # Base case
    if len(s) == 1:
        out = [s]
        
    else:
        
        #for every letter in string
        for i, letter in enumerate(s):
            
            # For every permutation resulting from step 2 and 3
            for perm in permute(s[:i] + s[i + 1:]): 
                
                print('current letter is ', letter)
                print('perm is: ', perm)
                
                # Add it to the output
                out += [letter + perm]
                
    return out
```

```python
permute('abc')
```
