```python
def rev_word1(s):
    return " ".join(reversed(s.split()))

#Or

def rev_word2(s):
    return " ".join(s.split()[::-1])
```

```python
rev_word1('Hi,  are you ready')
```

```python
def rev_word3(s):
    
    words = []
    length = len(s)
    space = [' ']
    
    i = 0 #index tracker
    
    while i < length:
        
        if s[i] not in space:
            word_start = i
            
            while i < length and s[i] not in space:
                 i += 1
            words.append(s[word_start : i])
            
        i += 1
        
    return " ".join(reversed(words))
```

```python
rev_word3(' space   before')
```
